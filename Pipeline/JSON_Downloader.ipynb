# Download Comments

To acquire the VOD comments I had to create a list of Twitch VOD IDs to look at. This list was generated by copying the ID from the URLs of all the relevant VODs. It has been annotated so that videos split up over multiple parts could have their comments appended. The range of dates used in this VOD list spans from July 27, 2015 to August 24, 2017. This list of IDs is then fed to a Python script, [rechat-dl](https://github.com/KunaiFire/rechat-dl) which output a JSON file of comments for each video.

```python
with open(r'VOD_ID.txt') as f:
    file = f.read()
shows = file.split('\n') #split into every show
shows
```

```python
import subprocess
for VOD in shows:
    subprocess.call(['python', 'rechat-dl.py', VOD])
```
