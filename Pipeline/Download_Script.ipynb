Message looks like it is the comment.

```python
#When running, replace 1 with row number
message_df = rechat_df[['message']]
message_df.head()
#This format will print out the comment from a row in the df, where 1 is the row of the comment, and cannot be 0
print(message_df.loc[1][0]['body'])
```

```python
commenter_df = rechat_df[['commenter']]
commenter_df.head()
print(commenter_df.loc[1][0]['display_name'])
```

```python
len(commenter_df)
len(message_df)
```

```python
#First 100 comments. Twitch recently allowed VOD comments, which is apparent.
for x in message_df.index[:100]: #For every number in index
    if x != 0: #Skip first index
        print(commenter_df.loc[x][0]['display_name'], ':', message_df.loc[x][0]['body'])
```

Lets make a DataFrame of comments

```python
comments = []
for x in message_df.index: #For every number in index
    if x != 0: #Skip first index
        comments.append([commenter_df.loc[x][0]['display_name'], message_df.loc[x][0]['body']])
        
comment_df = pd.DataFrame(comments)
comment_df.head()
```
